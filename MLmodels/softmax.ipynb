```python
import sklearn
import numpy as np
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
from softmax import SoftmaxRegression
%matplotlib inline
```

```python
dataset = load_iris()
features = pd.DataFrame(dataset.data,columns=dataset.feature_names)
target = pd.DataFrame(dataset.target)

sns.heatmap(features.corr(),cmap='Blues',annot=True)
```

```python
X = np.array(features)
y = np.array(target)

index = np.linspace(0,149,150,dtype=int)
np.random.shuffle(index)

X_train = X[index[0:100],:]
y_train = y[index[0:100],:]

X_test = X[index[100:],:]
y_test = y[index[100:],:]
```

```python
model = SoftmaxRegression()
model.fit(X_train,y_train,10000,0.1)
```

```python
model.weights
```

```python
model.score(X_train,y_train) , model.score(X_test,y_test)
```
