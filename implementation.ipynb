```python
import numpy as np
from linear_reg import *
from logistic_reg import *
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_california_housing, load_breast_cancer
```

# Linear Regression

```python
dataset = fetch_california_housing()
features = dataset['data']
targets = dataset['target'].reshape(-1,1)

split_size = 0.75

X_train, X_test = features[:int(split_size*features.shape[0])], features[int(split_size*features.shape[0]):]
y_train, y_test = targets[:int(split_size*features.shape[0])], targets[int(split_size*features.shape[0]):]
```

```python
lr = linear_reg()
lr.fit(X_train,y_train)
print("Weights are ",lr.weights.reshape(-1,))
print("Bias is ",lr.bias.reshape(-1,))

plt.title("Loss v/s Iteration")
plt.xlabel("Iteration")
plt.ylabel("Loss function")
plt.plot(lr.loss_history)
```

# Logistic Regression

```python
dataset = load_breast_cancer()
features = dataset['data']
targets = dataset['target'].reshape(-1,1)

split_size = 0.75

X_train, X_test = features[:int(split_size*features.shape[0])], features[int(split_size*features.shape[0]):]
y_train, y_test = targets[:int(split_size*features.shape[0])], targets[int(split_size*features.shape[0]):]
```

```python
lr = logistic_reg()
lr.fit(X_train,y_train)
print("Weights are ",lr.weights.reshape(-1,))
print("Bias is ",lr.bias.reshape(-1,))
print("Accuracy is", lr.accuracy(lr.prediction(X_test), y_test))

plt.title("Loss v/s Iteration")
plt.xlabel("Iteration")
plt.ylabel("Loss function")
plt.plot(lr.loss_history)
```
